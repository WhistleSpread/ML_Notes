### 构建一个简单的手写数字识别神经网络
输入层为784(28*28)，输出层为10,没有隐藏层

```python
import tensorflow as tf
from tensorflow.examples.tutorials import input_data
import numpy as np
```

```python
# 载入数据集，使用input_data这样的方法, 执行这个方法会直接给我们下载数据集
mnist = input_data.read_data_sets("MNIST_data", one_hot=True)
```

```python
# 我们训练模型的时候，不是一张一张的喂，而是一个批次一个批次的喂入，这个批次的大小可以自己定义
# 一般来讲，batch_size越大，得到的精确度越低，batch_size越小，精确度越高，不过训练起来也越慢


# batch_size = 100
batch_size = 50

# 计算一共有多少个批次
n_batch = mnist.train.num_examples // batch_size

# 定义两个placeholder
# 一个批次一个批次的传入到placeholder里面

x = tf.placeholder(tf.float32, [None, 784])
y = tf.placeholder(tf.float32, [None, 10])



# 创建一个简单的神经网络，只有输入层和有输出层， 没有隐藏层；
# W = tf.Variable(tf.zeros([784, 10]))
# b = tf.Variable(tf.zeros([10]))

# W = tf.Variable(tf.random_normal([784, 10], stddev=0.314, mean=0, seed=1))
# b = tf.Variable(tf.random_normal([10], stddev=0.314, mean=0, seed=1))


# # 预测值可以使用softmax函数, 通过softmax 将输出的信号转化为概率值
# prediction = tf.nn.softmax(tf.matmul(x, W) + b)

# 增加隐藏层，隐藏层设置为10

W1 = tf.Variable(tf.random_normal([784, 71], stddev=0.314, mean=0, seed=1))
W2 = tf.Variable(tf.random_normal([71, 10], stddev=0.314, mean=0, seed=1))
b1 = tf.Variable(tf.random_normal([71], stddev=0.314, mean=0, seed=1))
b2 = tf.Variable(tf.random_normal([10], stddev=0.314, mean=0, seed=1))

# 中间层的激活函数为tanh
a = tf.nn.tanh(tf.matmul(x, W1) + b1)
m = tf.nn.tanh(tf.matmul(a, W2) + b2)

# a = tf.nn.relu(tf.matmul(x, W1) + b1)
# m = tf.nn.relu(tf.matmul(a, W2) + b2)

# a = tf.matmul(x, W1) + b1
# m = tf.matmul(a, W2) + b2

# 预测值可以使用softmax函数, 通过softmax 将输出的信号转化为概率值
prediction = tf.nn.softmax(m)
# # prediction = tf.nn.relu(m)
# prediction = tf.nn.tanh(m)
# # prediction = tf.nn.sigmoid(m)

# 这里又一个非常重要的地方，这里一定要加reduce_mean 这个地方非常的重要，不加reduce_mean,得到的结果就很奇怪
loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels = y, logits = m))
# loss = tf.reduce_mean(tf.square(prediction - y))
train_step = tf.train.GradientDescentOptimizer(0.1).minimize(loss)

# 考虑使用其他的优化器来进行优化
# train_step = tf.train.GradientDescentOptimizer(0.01).minimize(loss)
init = tf.global_variables_initializer()

# 定义一下求准确率的方法：比较两个参数，如果一样的就返回True, 如果不一样就返回False
# correcct_prediction 就是一个list 
# tf.argmax(y, 1) 是求list y中最大的值是在哪一个位置
# Returns the index with the largest value across dimensions of a tensor. 
# Args: 
# input: A Tensor. Must be one of the following types: float32, float64, int64, int32, uint8, int16, int8, complex64, qint8, quint8, qint32. 
# dimension: A Tensor of type int32. int32, 0 <= dimension < rank(input). Describes which dimension of the input Tensor to reduce across. For vectors, use dimension = 0. 
# name: A name for the operation (optional). 
# Returns: A Tensor of type int64.
# dimension=0 按列找 
# dimension=1 按行找

correct_prediction = tf.equal(tf.argmax(y, 1), tf.argmax(prediction, 1))

# 求准确率, tf.cast 就是类型转换
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

with tf.Session() as sess:
    sess.run(init)
    for epoch in range(100):
        for batch in range(n_batch):
            # 在每一个批次中，也就是100张图片, 使用next_batch获得在这个batch中的x和y
            # 图片的像素数据保存在batch_xs中，图片的标签保存在batch_ys中
            # 下次运行的时候会自动的导入next_batch 中的数据
            batch_xs, batch_ys = mnist.train.next_batch(batch_size)
            sess.run(train_step, feed_dict={x:batch_xs, y:batch_ys})
            
        # 训练了一个周期之后，打印出准确率，看一下准确率的变化
        # 喂入的是测试集中的图片和测试集图片的标签
        acc = sess.run(accuracy, feed_dict={x:mnist.test.images, y:mnist.test.labels})
        print("Iter", str(epoch), "Testing Accuracy = ", str(acc))
```
