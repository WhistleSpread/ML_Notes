### 一个非线性模型

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

#### 人为生成我们需要的数据
使用numpy 生成200个随机点, 两百个点均匀分布到[-0.5, 0.5]区间之间, 并且为这个增加一个维度，也就是200行1列的数据,如果不增加[:,np.newaxis]的话，得到的就是一个list； np.newaxis的作用就是在原来的数组上增加一个维度。

np.array([1,2,3])[np.newaxis,:] 得到的结果和 a = np.array([1,2,3])[:,np.newaxis] 得到的结果不同

```python
x_data = np.linspace(-0.5, 0.5, 200)[:,np.newaxis]

# 生成噪音，这里x_data.shape 表示噪音的shape和x_data是一样的
noise = np.random.normal(0, 0.02, x_data.shape)
# 拟合的函数的形状是 y=x^2
y_data = np.square(x_data) + noise
```

### 构建一个神经网络，输入层为1， 隐藏层为10， 输出层为1

```python
# 定义两个placeholder 第一个参数是类型，第二个参数是形状,行数不确定，列数是1列
# placeholder的形状是根据样本的形状来定义的
x = tf.placeholder(tf.float32, [None, 1])
y = tf.placeholder(tf.float32, [None, 1])

# 定义输入层
# 定义一个1行10列的变量
# 偏置值都初始化为0，偏置值也是1行10列
# 中间层L1是通过tanh激活函数得到

Weights_L1 = tf.Variable(tf.random_normal([1, 10]))
biases_L1 = tf.Variable(tf.zeros([1, 10]))
Wx_plus_b_L1 = tf.matmul(x, Weights_L1) + biases_L1
L1 = tf.nn.tanh(Wx_plus_b_L1)


# 定义神经网络输出层
# 中间层是10个神经元，输出层是1个神经元，定义L2的权值是10行1列
# 因为输出只有1个，所以输出值的偏置值只有1个

Weights_L2 = tf.Variable(tf.random_normal([10, 1]))
biases_L2 = tf.Variable(tf.zeros([1, 1]))
Wx_plus_b_L2 = tf.matmul(L1, Weights_L2) + biases_L2
prediction = tf.nn.tanh(Wx_plus_b_L2)


# 定义loss function 以及定义训练方法
loss = tf.reduce_mean(tf.square(y - prediction))
train_step = tf.train.GradientDescentOptimizer(0.1).minimize(loss)
```

##### 开始训练
#### 训练了2000次之后，假设模型训练好了，喂入x_data 就可以得到预测的结果prediction

```python
with tf.Session() as sess:
#     print(x_data)
#     print(noise)
    # 初始化所有变量，喂入数据并训练2000次
    sess.run(tf.global_variables_initializer())
    for _ in range(2000):
        sess.run(train_step, feed_dict={x:x_data, y:y_data})
    
    
    prediction_value = sess.run(prediction, feed_dict={x:x_data})
    
    plt.figure()
    # 用散点图画出实际的值
    plt.scatter(x_data, y_data)
    
    # 用实线画出预测的值和x_data的曲线
    plt.plot(x_data, prediction_value, 'r-', lw=5)
    plt.show()
```
